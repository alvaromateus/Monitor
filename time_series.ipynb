```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import datetime
from matplotlib.pylab import rcParams

import pandas as pd

# 03/13/2020 14:58:08.719
dateparse = lambda dates: pd.datetime.strptime(dates, '%m/%d/%Y %H:%M:%S.%f')

# abre aquivo csv para fazer parse da data
data = pd.read_csv('log.csv', parse_dates=['data_hora'], index_col='data_hora', date_parser=dateparse)

# imprime tipos de dados e primeiros registros
print '\nData Types:'
print data.dtypes
print data.head()
```

```python
# gera variaveis com dados dos percentuais de consumo de memoria e cpu
ts_memoria = data['memoria']
ts_cpu = data['cpu']
```

```python
# plota gráficos dos dados de consumo de memória
plt.plot(ts_memoria)
plt.title("Consumo de Memoria")
```

```python
# plota gráficos dos dados de consumo de memória
plt.plot(ts_cpu)
plt.title("Consumo de CPU")
```

```python
# plota gráficos dos dados de consumo de memória e CPU
plt.plot(ts_cpu, label="Percentual consumo CPU")
plt.plot(ts_memoria, label="Percentual consumo memoria")
plt.title("Consumo de Memoria  e CPU")
plt.legend()
```

```python
# calcula media movel de 1000 periodos
moving_avg = ts_memoria.rolling(1000).mean()

#plota grafico do consumo de memoria
plt.plot(ts_memoria,label="Grafico 1000 periodos")
plt.plot(moving_avg, color='red', label='Media Movel 1000 periodos')
plt.title("Consumo de Memoria - Media movel")
plt.legend()
```

```python
# calcula media movel de 1000 periodos
moving_avg = ts_cpu.rolling(1000).mean()

#plota grafico do consumo de memoria
plt.plot(ts_cpu,label="Grafico total")
plt.plot(moving_avg, color='red', label='Media Movel 1000 periodos')
plt.title("Consumo de CPU - Media movel")
plt.legend()
```
